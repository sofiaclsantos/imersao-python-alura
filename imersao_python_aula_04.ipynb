```python
import pandas as pd
# import das bibliotecas de visualizacao de dados e da API do yahoo para dados do mercado financeiro
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# criacao de planilha com dados da petrobras durante o ano de 2023
dados = yf.download('PETR4.sa', start='2023-01-01', end='2023-12-31')
dados
```

```python
# renomeando as colunas
dados.columns = ['Abertura', 'Maximo', 'Minimo', 'Fechamento', 'Fech_Ajust', 'Volume']
```

```python
# renomeando o eixo de data
dados = dados.rename_axis('Data')
dados
```

```python
# criando o grafico de variacao utilizando os dados da coluna de fechamento
dados['Fechamento'].plot(figsize=(10,6))
plt.title('Variação do preço por data', fontsize=16)
plt.legend(['Fechamento'])
```

```python
df = dados.head(60).copy()
# convertendo o indice em uma coluna de data
df['Data'] = df.index
# convertendo as datas para o formato numerico do matplotlib (p/ que a plotagem seja feita corretamente)
df['Data'] = df['Data'].apply(mdates.date2num)
```

```python
df
```

Gráfico de candlesticks (construção total)

```python
fig, ax = plt.subplots(figsize=(15, 8))

# largura dos candles
width = 0.7

#  cor verde para fechamento>abertura, caso contrario vermelho
for i in range(len(df)):
    if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
        color = 'green'
    else:
        color = 'red'

    # desenhando a linha vertical/sombra (indica max e min do dia)
    # `ax.plot`  desenha uma linha vertical.
    # no argumento de ax.plot (cria a linha do candle) ficam os dois primeiros valores do ponto x, e depois a altura (min e max)
    ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
            [df['Minimo'].iloc[i], df['Maximo'].iloc[i]],
            color=color,
            linewidth=1)

    # plotando o corpo o candle
    # `ax.add_patch` adiciona um retangulo ao grafico
    ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                               width,
                               abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                               facecolor=color))

 # medias moveis
df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] = df['Fechamento'].rolling(window=14).mean()

# plotando medias moveis
ax.plot(df['Data'], df['MA7'], color='orange', label='Média móvel de 7 Dias')  # 7 dias
ax.plot(df['Data'], df['MA14'], color='purple', label='Média móvel de 14 Dias')  # 14 dias
ax.legend()

# formatando o eixo x para mostrar as datas
ax.xaxis_date() # indica que as datas estao sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) # colocando no formato de data

# titulos e rotulos
plt.title("Gráfico de Candlestick - PETR4.SA")
plt.xlabel("Data")
plt.ylabel("Preço")

# insere grade
plt.grid(True)

# exibe o grafico
plt.show()
```

Gráfico dinâmico com subplots

```python
# criando subplots para ter um grafico de candlestick e um de barras
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1,
                    subplot_titles=('Candlesticks', 'Volume Transacionado'),
                    row_width=[0.2, 0.7])

# grafico de candlestick
fig.add_trace(go.Candlestick(x=df.index,
                             open=df['Abertura'],
                             high=df['Maximo'],
                             low=df['Minimo'],
                             close=df['Fechamento'],
                             name='Candlestick'),
                             row=1, col=1)

# medias moveis do candlestick
fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA7'],
                         mode='lines',
                         name='MA7 - Média Móvel 7 Dias'),
                         row=1, col=1)

fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA14'],
                         mode='lines',
                         name='MA14 - Média Móvel 14 Dias'),
                         row=1, col=1)

# graficos de barras (volume de transacoes)
fig.add_trace(go.Bar(x=df.index,
                     y=df['Volume'],
                     name='Volume'),
                     row=2, col=1)

# configuracoes visuais da figura
fig.update_layout(yaxis_title='Preço',
                  xaxis_rangeslider_visible=False,  # desativa o range slider
                  width=1100, height=600)

# exibe grafico
fig.show()
```
